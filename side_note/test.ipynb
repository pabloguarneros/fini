```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
ts = np.arange(0,10,1)
above_1 =   lambda x: math.e**(1.2*x)
below_1 = lambda x: math.e**(0.008*x)
plt.plot(ts, below_1(ts))
```

```python
rs = np.arange(0,100,1)
test_denom = lambda r: 40/(math.e**(.25*r))
plt.plot(rs, test_denom(rs))
```
